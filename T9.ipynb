```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
plt.style.use('ggplot')
%matplotlib inline
```

**Adicionando as distribuições das taxas de corrosão obtidas.**

```python
data = {'A-FI42(14X6)':[30.03, 61.18, 47.71, 95.98, 79.88, 43.30, 101.94, 87.11],
        'B-FI42(20X5)':[65.18, 52.65, 31.71, 24.16, 21.09, 50.45, 11.79, 35.02],
        'C-FI27(20X6)':[70.43, 54.59, 33.17, 95.84, 105.69, 98.67, 83.40, 74.56],
        'D-FI18(Teta14)':[15.01, 23.60, 13.21, 11.43, 15.69, 15.95, 14.57, 15.26],
        'E-FI09(Zeta10)':[154.55, 138.34, 258.64, 250.88, 138.86, 208.59, 241.24, 148.81]}
df = pd.DataFrame(data=data)
df
```

```python
df.describe()
```

```python
sns.set_style("whitegrid")
fig2, ax2 = plt.subplots(figsize=[18, 14])
sns.set(font_scale=1.5)
sns.boxplot(data=df, orient='v', ax=ax2, width=0.5, linewidth=2, palette='muted' )
A = mpatches.Patch(color='cornflowerblue', label='A-FI42(14x6)')
B = mpatches.Patch(color='chocolate', label='B-FI42(20x5)')
C = mpatches.Patch(color='mediumseagreen', label='C-FI27(20x6)')
D = mpatches.Patch(color='firebrick', label='D-FI18(Teta14)')
E = mpatches.Patch(color='darkorchid', label='E-FI09(Zeta10)')
plt.legend(handles=[A, B, C, D, E], loc='upper left', title='Samples', title_fontsize=25, framealpha=1, shadow=True, fontsize='large')
plt.show()
```

**Comentários:**
- As amostras A,B e C tem distribuições muito próximas em magnitude e disperções parecidas, com exceção das amostras D e E.

- As diferenças entres as amostras pode ter sido causadas devido ao formato das peças, as amostras D e E apresentam sulcos em suas estruturas que podem ajudar no aumento da taxa de corrosão.


# **1) A-FI42(14x6) x B-FI42(20x5) -> (9A x 9B)**

**Teste de Hipótese Bilateral - Variáveis Indenpendentes**

```python
stats.ttest_ind(df['A-FI42(14X6)'],df['B-FI42(20X5)'], equal_var=False)
```

**p-valor = 1.56142e-02, muito baixo, hipótese nula rejeitada.**

```python
#fig2, ax2 = plt.subplots(figsize=[10, 7])
#fig2.subplots_adjust(top=0.9)
#plt.suptitle('Distribuições A, B', fontsize=14, fontweight='bold')
p1 = sns.kdeplot(df['A'], shade=True, color='r')
p1 = sns.kdeplot(df['B'], shade=True, color='b')
plt.title('Distribuições')
plt.show()
```

```python
sns.kdeplot(df['B'],df['A'], shade=False, cmap='coolwarm',cbar=True, shade_lowest=True, n_levels=18)
plt.title('A x B')
plt.show()
```
